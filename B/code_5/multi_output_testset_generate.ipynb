```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
table_path = "../data/data_num_325_fea_30_rf_multi_labels.csv"
#table_path = "../data/rf_top_30_data_preprocess_45_multi_labels.csv"
data = pd.read_csv(table_path)
#data.to_csv("../data/data_preprocess_only_45_S_RON_loss.csv")
data.head()
```

```python
#"Product_RON"
y = data[["Product_S_content","Product_RON","Product_RON_loss"]]
y.head()
```

```python
X = data.drop("Product_S_content",1)
X = X.drop("Product_RON",1)
X = X.drop("Product_RON_loss",1)
X.head()
```

```python
#Feature Scaling - Mean to 0, variance to 1
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X.head()
```

```python
# Split data into separate training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 45)
```

```python
from sklearn.externals import joblib
model_path_name = "../model/multi_output_ridge_new.m"
clf = joblib.load(model_path_name)
y_test_predict= clf.predict(X_test)   # Make predictions and save it in dict under key: name
y_mse = mean_squared_error(y_test, y_test_predict)
print(y_mse)
```

```python
#特征重要度
features = list(X_test.columns)
importances = pd.DataFrame(clf.coef_,columns=X.columns,index=y.columns)
importances
```

```python
#按重要程度排序
importances.loc['abs'] = importances.loc['Product_RON_loss'].abs()
importances = importances.sort_values(by='abs',axis=1,ascending=False)
importances
```

```python
importances.columns
```

```python
#构建操作矩阵
operate = importances.T
operate = operate.drop('Product_RON',1)
#operate = operate.drop('Recycle_S')
operate = operate.drop('Raw_RON')
#operate = operate.drop('Recycle_C')
#operate = operate.drop('Raw_Saturated_hydrocarbon')
operate
```

```python
#要使Product_RON_loss降低，正系数的特征要调低，负系数的特征要调高
#要保持Product_S_content不升高，要选择系数与Product_RON_loss同号的，保证一起下降

for column in operate.index:
    if operate.loc[column,'Product_RON_loss']>0 and operate.loc[column,'Product_S_content']>0:
        operate.loc[column,'operate']='调低'
    elif operate.loc[column,'Product_RON_loss']<0 and operate.loc[column,'Product_S_content']<0:
        operate.loc[column,'operate']='调高'
    else:
        operate = operate.drop(column)
operate
```

```python
range_table_path = "../data/data_range.csv"
feature_range = pd.read_csv(range_table_path,index_col = 0)
feature_range
```

```python
feature_range['S-ZORB.LT_1501.DACA']
```

```python
for column in operate.index:
    #print(column)
    operate.loc[column,'delta']=feature_range.loc['delta',column]
    if operate.loc[column,'operate']=='调高':
        operate.loc[column,'target']=feature_range.loc['r2',column]
    elif operate.loc[column,'operate']=='调低':
        operate.loc[column,'target']=feature_range.loc['r1',column]
operate
```

```python
operate.loc['S-ZORB.LT_1501.DACA','delta'] = 0.1
operate
```

```python
#Load data
data_table_path = "../data/data_num_325_fea_30_rf_multi_labels.csv"
#data_table_path = "../data/rf_top_30_data_preprocess_45_multi_labels.csv"
raw_data = pd.read_csv(data_table_path)
raw_data.head()
```

```python
data_133 = raw_data.iloc[132:133,:]
data_133.index = data_133.index - 132
data_133
```

```python
#Compute target:
data_133_X = data_133.drop('Product_S_content',1)
data_133_X = data_133_X.drop('Product_RON',1)
data_133_X = data_133_X.drop('Product_RON_loss',1)
data_133_X
```

```python
data_133_target = data_133_X.copy()
data_133_target
```

```python
for column in operate.index:
    data_133_target.loc[0,column] = operate.loc[column,'target']
data_133_target
```

```python
#Insert data from raw data to target
data_133_test = data_133_X.copy()
for i in range(100):
    l = len(data_133_test)
    Flag_change = 0
    new = data_133_test.iloc[l-1:,:].copy()
    new.index = new.index - new.index
    for column in operate.index:
        #print(operate.loc[column,'operate'])
        if operate.loc[column,'operate'] =='调高':
            if float(data_133_test.at[l-1,column]) < float(data_133_target.at[0,column]):
                Flag_change = 1
                new.loc[0,column] = float(data_133_test.at[l-1,column]) + float(operate.at[column,'delta'])
        elif operate.loc[column,'operate'] =='调低':
            if float(data_133_test.at[l-1,column]) > float(data_133_target.at[0,column]):
                Flag_change = 1
                new.loc[0,column] = float(data_133_test.at[l-1,column]) - float(operate.at[column,'delta'])
    if Flag_change == 0:
        break
    data_133_test=data_133_test.append(new,ignore_index=True)
data_133_test
```

```python
data_133_test.to_csv("../data/multi_output_testset_new.csv",index=None)
```
