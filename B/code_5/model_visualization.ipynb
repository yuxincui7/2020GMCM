```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
table_path = "../data/multi_output_testset_new.csv"
X_test = pd.read_csv(table_path)
X_test.head()
```

```python
#Feature Scaling - Mean to 0, variance to 1
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns = X_test.columns)
X_test.head()
```

```python
from sklearn.externals import joblib
model_path_name = "../model/multi_output_ridge_new.m"
clf = joblib.load(model_path_name)
y_test_predict = clf.predict(X_test)   # Make predictions and save it in dict under key: name
y_test_predict = pd.DataFrame(y_test_predict,columns=["Product_S_content","Product_RON","Product_RON_loss"])
y_test_predict
```

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = y_test_predict[['Product_S_content']]
y2 = y_test_predict[['Product_RON']]
y3 = y_test_predict[['Product_RON_loss']]
plt.plot(y1,label='Product_S_content')
#plt.plot(y2,label='Product_RON')
plt.plot(y3,label='Product_RON_loss')
plt.legend()
```
