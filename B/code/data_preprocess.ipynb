```python
old_table = pd.read_excel(old_table_path)
old_table.head()
table = old_table.iloc[2:]
fea_name_add = ["Raw_S_content","Raw_RON","Raw_Saturated_hydrocarbon","Raw_Olefins","Raw_Aromatics","Raw_Bromine_value","Raw_Density","Product_S_content","Product_RON","Product_RON_loss","Spent_C","Spent_S","Recycle_C","Recycle_S"]
fea_name = old_table.iloc[0].values.tolist()
#print(fea_name)
fea_name[2:16] = fea_name_add
print(fea_name)
table.columns = fea_name
table.set_index(["样本编号"], inplace=True)
table.head()
```

```python
for col in col_name[1:]:
    #print(col)
    old_data = table.loc[285,col]
    #print(old_data)
    new_data = data_285_mean.loc[0,col]
    #print(new_data)
    flag = 0
    if new_data != old_data:
        table.loc[285,col] = new_data
        flag = 1
        print(col, old_data, new_data)
if flag == 0:
    print("Same data!")
```
