```python
import re
r1_list = []
r2_list = []
for i in range(0, len(data_range['range'])):
    r1 = 0.0
    r2 = 0.0
    flag1 = 1
    flag2 = 1
    a = re.split(r'-', data_range['range'][i])
#     print('a', a)
    if len(a) == 3:
        flag1 = -1
        r1 = a[1]
        r2 = a[2]
    elif len(a) == 4:
        flag1 = -1
        flag2 = -1
        r1 = a[1]
        r2 = a[3]
        r2 = r2[0:-1]
    else:
        r1 = a[0]
        r2 = a[1]
#     print('*****' , i)
#     print('r1', r1)
#     print('r2', r2)
    if r2.startswith('(') or r2.startswith('（'):
        r2 = r2[1:-1]
#         print(r2)
    r2 = flag2 * float(r2)
    r1 = flag1 * float(r1)
#     data_range.loc[i,'r1'] = r1
#     data_range.loc[i,'r2'] = r2
    r1_list.append(r1)
    r2_list.append(r2)
data_range['r1'] = r1_list
data_range['r2'] = r2_list
```
